```python
# import csv file
df = pd.read_csv(r'C:\Users\akash\Downloads\Python_Diwali_Sales_Analysis-main\Python_Diwali_Sales_Analysis-main\Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
sns.countplot(x = 'Gender',data = df)
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
#From the analysis of bar graph we can say that most the buyers are women and even the purchasing power of women is greater than men
```

```python
ax = sns.countplot(x = 'Age Group',data = df,hue = 'Gender' )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python

df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
#amount vs age group
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y= 'Amount',data = sales_age)
```

```python
#Analaysing the above graphs we can say that most of the buyers are age group between 26-35 women
```

```python
df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)
```

```python
#analysing sales for top 15 state
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=sales_state,x='State',y='Orders')


```

```python
df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending = False).head(15)
```

```python
#amount vs sales state
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=sales_state,x='State',y='Amount')

```

```python
#from thwe analysis we can say that top sates for sales and amount are uttar pradesh,maharastha,karnataka in order
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(10,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
#from the graphs we can say that most of the buyers are married(women) and they have greater purchasing power
```

```python
#Occu[ation
ax =sns.countplot(x='Occupation',data = df)
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# now we analysis amount vs occupation
sales_Occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x ='Occupation',y='Amount',data=sales_Occupation)

```

```python
#from analysising of graphs we can say that sales occupation and purchasing amount is greater in it sector health care aviation
```

```python
#product catagory
ax = sns.countplot(data = df,x='Product_Category')
sns.set(rc={'figure.figsize':(15,7)})
for bars in ax.containers:
    ax.bar_label(bars)


```

```python
#product catagory vs amount
Product_Category =df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(38,17)})
sns.barplot(x='Product_Category',y='Amount',data =Product_Category)


```

```python
#From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```
